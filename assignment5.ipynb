```python
features = data.drop('DEATH_EVENT', axis=1)  # Assuming 'target_column' is your target variable
target = data['DEATH_EVENT']
```

```python
pca = PCA(n_components=2)  # Here, we're keeping 2 components
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
pca = PCA(n_components=4)
pca.fit(scaled_features)

pca_X = pca.transform(scaled_features)
```

```python
PC1 = pca_X[:, 0]
PC2 = pca_X[:, 1]
PC3 = pca_X[:,2]
PC4 = pca_X[:,3]
```

```python

plt.scatter(PC1, PC2, PC3, PC4,alpha=0.8)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Heart Dataset PC1 vs PC2')
```

```python
plt.scatter(PC3, PC4,alpha=0.8)
plt.xlabel('Principal Component 3')
plt.ylabel('Principal Component 4')
plt.title('PCA of Heart Dataset PC3 vs PC4')
```
